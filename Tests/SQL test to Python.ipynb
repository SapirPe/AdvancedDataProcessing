### Q2

```python
#a: לכל ספק מספר מוצרים שהוא מספק
prodOfSup_2024_b = pd.merge(SuppliersDate[['supplierID', 'companyName']],ProductsDate[['productID', 'supplierID']], how='inner',on='supplierID')
prodOfSup_2024_b=prodOfSup_2024_b.groupby(['supplierID','companyName']).count()
prodOfSup_2024_b=pd.DataFrame(prodOfSup_2024_b).reset_index()
prodOfSup_2024_b.columns = ['supplierID', 'companyName','countProducts']
prodOfSup_2024_b.head(3)
```

```python
#b: סכום כמויות שנמכרו סה"כ מכל המוצרים שלו

help_merge = pd.merge(SuppliersDate[['supplierID', 'companyName']],ProductsDate[['productID', 'supplierID']], how='inner',on='supplierID')
totSoldProd_2024_b = pd.merge(help_merge,OrderDetailsDate[['productID','quantity']], how='inner',on='productID')

totSoldProd_2024_b = totSoldProd_2024_b[['supplierID', 'companyName','quantity']]
totSoldProd_2024_b=totSoldProd_2024_b.groupby(['supplierID','companyName']).sum()
totSoldProd_2024_b=pd.DataFrame(totSoldProd_2024_b).reset_index()
totSoldProd_2024_b.columns = ['supplierID', 'companyName','sumSoldProd']
totSoldProd_2024_b.head(3)
```

```python
#c: סך פדיון מכל המוצרים ביחד

#הוספת עמודות:
help_OD = OrderDetailsDate[['productID','quantity','unitPrice']]
help_OD['revenue'] = help_OD['quantity'] * help_OD['unitPrice']

#מיזוג+ סלקט
help_merge = pd.merge(SuppliersDate[['supplierID', 'companyName']],ProductsDate[['productID', 'supplierID']], how='inner',on='supplierID')
totRev_2024_b = pd.merge(help_merge,help_OD, how='inner',on='productID')

totRev_2024_b = totRev_2024_b[['supplierID', 'companyName','revenue']]


totRev_2024_b=totRev_2024_b.groupby(['supplierID','companyName']).sum()
totRev_2024_b=pd.DataFrame(totRev_2024_b).reset_index()
totRev_2024_b.columns = ['supplierID', 'companyName','sumRevenue']
totRev_2024_b.head(3)
```

```python
#d: תרומה מההכנסות
#totRev_2024_b['sumRevenue'] #סך פדיון של הספק מהמכירות

#הוספת עמודות:
help_data = OrderDetailsDate[['quantity','unitPrice']]
help_data['organization_total_revenue'] = help_data['quantity'] * help_data['unitPrice']

org_tot_rev=help_data['organization_total_revenue'].sum() #SELECT בS משתנה ששומר את סכום הפדיון הכללי בארגון < מקביל לתת שאילתה  



#הוספת יחס התרומה לטבלה מסעיף ג
totRev_2024_b['%Teruma'] = (totRev_2024_b['sumRevenue']/org_tot_rev)*100
totRev_2024_b.head(3)
```

```python
#e
#איחוד של כל הטבלאות:

help_merge= pd.merge(prodOfSup_2024_b,totSoldProd_2024_b[['supplierID','sumSoldProd']], how='inner', on='supplierID' )
result = pd.merge(help_merge,totRev_2024_b[['supplierID','sumRevenue','%Teruma']], how='inner', on='supplierID' )
```

```python
#ORDER BY
result =result.sort_values(by='sumRevenue', ascending=False)
result.head(3)
```

```python
#WHERE סינון ב  
con1 = CustomersDate['customerID']=='ALFKI'
customers_help= CustomersDate[['customerID']][con1] #רק הלקוח אלפקי

help_merge=pd.merge(customers_help, OrderDate[['orderID','customerID']], how='inner', on= 'customerID')
ALFKI_orders = help_merge['orderID'].nunique()

#--

con1 = CustomersDate['customerID']=='QUICK'
customers_help= CustomersDate[['customerID']][con1] #רק הלקוח קוויק

help_merge=pd.merge(customers_help, OrderDate[['orderID','customerID']], how='inner', on= 'customerID')
quick_orders = help_merge['orderID'].nunique()
minOrders=quick_orders-ALFKI_orders
minOrders
```

```python
#a: לכל מדינה את כמות הלקוחות וכמות ההזמנות בה
help_merge = pd.merge(CustomersDate[['country','customerID']], OrderDate[['orderID','customerID']], how='inner', on= 'customerID')
help_merge = help_merge.groupby('country')[['customerID','orderID']].nunique()
countCusAndOrders=pd.DataFrame(help_merge).reset_index()
countCusAndOrders.columns = ['country', 'countCustomers','countOrders']

#הבטחה שהמדינות יעמדו בתנאי של כמות ההזמנות
countCusAndOrders = countCusAndOrders[countCusAndOrders['countOrders']>minOrders]
countCusAndOrders.head(30)
```

```python
#b: לכל מדינה את ממוצע הכמויות בהזמנה
h_merge = pd.merge(CustomersDate[['country','customerID']], OrderDate[['orderID','customerID']], how='inner', on= 'customerID')
help_merge=pd.merge(h_merge, OrderDetailsDate[['orderID','quantity']], how='inner', on= 'orderID')
help_merge=help_merge[['country','quantity']]
averageQuan = help_merge.groupby('country')[['quantity']].mean()
averageQuan=pd.DataFrame(averageQuan).reset_index()
averageQuan.columns = ['country', 'averageQuanInOrder']
averageQuan.head(3)
```

```python
#c: לכל מדינה את כמות המוצרים השונים שרכשו
h_merge = pd.merge(CustomersDate[['country','customerID']], OrderDate[['orderID','customerID']], how='inner', on= 'customerID')
help_merge=pd.merge(h_merge, OrderDetailsDate[['orderID','productID']], how='inner', on= 'orderID')
help_merge=help_merge[['country','productID']]
diffProd = help_merge.groupby('country')['productID'].nunique()
diffProd=pd.DataFrame(diffProd).reset_index()
diffProd.columns = ['country', 'countDiffProductd']
diffProd.head(3)
```

```python
#d: איחוד לטבלה אחת סופית
m1 = pd.merge(countCusAndOrders,averageQuan, on= 'country')
m2 = pd.merge(m1,diffProd, on= 'country')
m2
```

```python
con1 = OrderDate['orderDate'].between('1996-07-07','1997-07-07')

help_Order = OrderDate[['employeeID','orderID','orderDate']][con1]
help_Order['month']=help_Order['orderDate'].dt.month

h_merge = pd.merge(EmployeeDate[['employeeID','lastName','firstName']],help_Order,on='employeeID')

help_OD = OrderDetailsDate[['orderID','quantity','unitPrice']]
help_OD['rev'] = help_OD['quantity']* help_OD['unitPrice']

help_merge = pd.merge(h_merge,help_OD[['orderID','rev']],on='orderID')

help_merge=help_merge[['employeeID','lastName','firstName','month','rev']]
avgRev = help_merge.groupby(['employeeID','lastName','firstName','month']).mean()
avgRev=pd.DataFrame(avgRev).reset_index()
avgRev.columns = ['employeeID','lastName','firstName','month','avgRevenue']
avgRev
```
